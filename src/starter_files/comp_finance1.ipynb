```python
def annualize_rets(r, periods_per_year=12):
    n_periods = r.shape[0]
    compounded_growth = (1+r).prod()
    return compounded_growth**(periods_per_year/n_periods)-1
```

```python
annualize_rets(rets_pct_change)
```

```python
def annualize_vol(r, periods_per_year=12):
    return r.std() * (periods_per_year**0.5)
```

```python
annualize_vol(rets_pct_change)
```

```python
# Raw Sharpe Ratio
annualize_rets(rets_pct_change) / annualize_vol(rets)
```

```python
wealth_index = (rets_pct_change + 1).cumprod()
start_date = wealth_index.index.min() - pd.DateOffset(months=1)
wealth_index = pd.concat([pd.Series([1], index=[start_date]), wealth_index])

previous_peaks = wealth_index.cummax()
drawdowns = (wealth_index - previous_peaks) / previous_peaks

# Scalars instead of Series
max_drawdown = drawdowns.min().min()
max_drawdown_date = drawdowns.stack().idxmin()[0]

plt.figure(figsize=(12,6))
plt.plot(wealth_index, label="Wealth Index")
plt.plot(previous_peaks, label="Previous Peaks")
plt.plot(drawdowns, label="Drawdowns")

plt.annotate(f"Max Drawdown: {max_drawdown:.2%}",
             xy=(max_drawdown_date, max_drawdown),
             xytext=(max_drawdown_date + pd.DateOffset(years=24), max_drawdown*0.5),
             arrowprops=dict(arrowstyle="->", lw=1),
             color="blue")

plt.legend()
plt.show()
```
